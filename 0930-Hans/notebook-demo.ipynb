# Title of investigation

Want to understand the equation $f(t) = \exp(-\alpha t) \cos(\omega t)$. Let's write it in Python:

```python
def f(t, alpha, omega):
    """Computes and returns exp(-alpha*t) * cos(omega*t)"""
    return exp(-alpha * t) * cos(omega * t)
```

We can execute the function for valuel of $\alpha$ and $\omega$:

```python
f(t=0.0, alpha=1, omega=10)
```

Although sometimes a plot is more instructive:

```python
def plot_f(alpha, omega):
    ts = linspace(0, 5, 400)    # 400 points in the interval [0, 5]
    ys = f(ts, alpha, omega)
    pylab.plot(ts, ys, '-')
```

```python
plot_f(alpha=0.1, omega=20)
```

```python
plot_f(alpha=0.1, omega=10)
```

```python
plot_f(alpha=0.5, omega=10)
```

It seems that $\alpha$ is the damping.
